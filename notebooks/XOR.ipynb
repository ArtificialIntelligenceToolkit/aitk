```python
inputs = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1],
]
targets = [0, 1, 1, 0]
```

```python
import aitk.networks
aitk.networks.__version__
```

```python
net = aitk.networks.SimpleNetwork(2, 2, 1, activation="sigmoid")
```

```python
net.take_picture()
```

```python
net.fit(
    inputs, targets, 
    epochs=3000, 
    batch_size=4, 
    report_rate=100, 
    accuracy=1.0, 
    patience=1,
    save=100,
)
```

```python
for epoch, weights in net.get_weights_from_history():
    net.set_weights(weights)
    net.predict_pca(inputs, scale=.5, colors=["r", "b", "b", "r"], sizes=300)
    input("Epoch %s" % epoch)
```
