```python
import aitk.robots as bots
import math
```

```python
w = bots.World(100, 100, scale=10)
w.watch(width=500)
robot = bots.Scribbler(50, 50, 0)
for angle in bots.utils.arange(0, 359, 360 / 12):
    x, y = bots.utils.rotate_around(0, 0, 7, -angle * math.pi / 180)
    robot.add_device(bots.RangeSensor((x, y), a=angle, width=30))
w.add_robot(robot)
w.draw()
```
