```python
from aitk.algorithms.ga import GeneticAlgorithm
import random
```

```python
class SumGA(GeneticAlgorithm):
    """
    An example of using the GeneticAlgorithm class to solve a particular
    problem, in this case finding strings with the maximum number of 1's.
    """
    def fitness(self, chromosome):
        """
        Fitness is the sum of the bits.
        """
        return sum(chromosome)
    
    def is_done(self):
        """
        Stop when the fitness of the the best member of the current
        population is equal to the maximum fitness.
        """
        return self.fitness(self.bestEver) == self.length
    
    def make_random_gene(self):
        """
        Function to generate a new random chromosome.
        """
        # Override this
        return random.choice([0, 1])
```

```python
ga = SumGA(20, 50) # Bit strings of length 20, population of size 50
bestFound = ga.evolve(
    generations=100, 
    crossover_rate=0.6, 
    mutation_rate=0.001, 
    elite_percent=0.1)
ga.plot_stats("Sum GA")
```
